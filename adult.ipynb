```python
#Feature ingineering
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
df.income = le.fit_transform(df.income)
df
```

```python
df['income'].value_counts()  #7841 ceux qui gagne un salaire <=50K et 24720 un salaire >50K
```

```python
# create a Python list of feature names
feature_cols = ['age', 'fnlwgt', 'education.num','capital.gain','capital.loss','hours.per.week']
X = df[feature_cols]
y = df['income']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=1)
```

```python
# Import DecisionTreeClassifier from sklearn.tree
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import mean_squared_error

# Instantiate a DecisionTreeClassifier 'dt' with a maximum depth of 6
dt = DecisionTreeClassifier(max_depth =6)

# Fit dt to the training set
dt.fit(X_train, y_train)

# Predict test set labels train set 
y_pred = dt.predict(X_train)
# Le MSE(Mean Squared Error)
mse_train = mean_squared_error(y_train,y_pred)
mse_train
y_t = dt.predict(X_test)
mse_test = mean_squared_error(y_test,y_t)
```

```python
from sklearn.linear_model import LogisticRegression
# instantiate a logisticRegression
model = LogisticRegression()
# Fit dt to the training set
model.fit(X_train,y_train)
# Predict test set labels
y_pred = model.predict(X_test)
```
